```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dfx = pd.read_csv('C:\\Users\\Ayus Das\\Practice\\Linear_X_Train.csv')
dfy = pd.read_csv('C:\\Users\\Ayus Das\\Practice\\Linear_Y_Train.csv')
dfxtest = pd.read_csv('C:\\Users\\Ayus Das\\Practice\\Linear_X_Test.csv')
x = dfx.values
y = dfy.values
xtest = dfxtest.values
xtest = xtest.reshape((-1,1))


x = x.reshape((-1,1))
y = y.reshape((-1,1))

X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
output = model.predict(X)
```

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

```python
model.score(X,Y)
```

```python
## Visualise

plt.scatter(X,Y,label='data')
plt.plot(X,output,color='orange',label='prediction')
plt.legend()
plt.show()
```
